```python
import matplotlib.pyplot as plt
import math
import numpy
import random
```

#### Geometric (lognormal) random walk

```python
x = range(0, 100)
y = 100 * numpy.cumprod([0.99 if random.uniform(0, 1) < 0.5 else 1.01 for i in x])
plt.plot(x,y)
plt.show()
```

#### Arithmetic (normal) random walk

```python
x = range(0, 100)
y = 100 + numpy.cumsum([-1 if random.uniform(0, 1) < 0.5 else 1 for i in x])
plt.plot(x,y)
plt.show()
```

#### Interest rates
If $r$ is annual IR, then $(1 + r)^n$ is interest rate in n years.

If there are m interest payments at rate $\frac{r}{m}$, than in one year we will have
$(1 + \frac{r}{m})^m$

in case of continuous compounding where $m \to \infty$, 

$(1 + \frac{r}{m})^{m t} \sim e^{r t}$

#### Forwards and Futures

##### Simple Futures
$F = S(t)e^{r(T-t)}$, $F$ - future price, $S(t)$ - spot price at time $t$
##### Commodity futures
$F = S(t)e^{(r+s-c)(T-t)}$, $s$ - storage cost, $c$ - convenience yield
##### FX futures
$F = S(t)e^{(r-r_f)(T-t)}$, where $r_f$ is interest received from foreign currency
##### Index futures
$F = S(t)e^{(r-q)(T-t)}$, where $q$ are stock dividend yield
